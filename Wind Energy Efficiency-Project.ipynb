```python
for col in df.columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    LB=Q1-1.5*(IQR)
    UB=Q3+1.5*(IQR)

    df=df[(df[col]>=LB) & (df[col]<=UB)]

Q1,Q3,IQR,LB,UB
```

```python
df['timestamp']=pd.to_datetime(df['timestamp'])
```
