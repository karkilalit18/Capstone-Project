```python
def detect_intent(text):
    text = text.lower()

    high_intent_words = [
        "want to try", "i want", "sign up", "signup", "register",
        "buy", "subscribe", "get started", "start using", "purchase"]

    product_words = [
        "price", "pricing", "plan", "plans", "cost", "feature", "features",
        "basic", "pro", "refund", "support"]

    greeting_words = ["hi", "hii", "hello", "hey"]

    if any(w in text for w in high_intent_words):
        return "high_intent"

    if any(w in text for w in product_words):
        return "product_query"

    if any(w in text for w in greeting_words):
        return "greeting"

    return "unknown"
```

```python
def retrieve_answer(query):
    q = query.lower()

    if "pricing" in q or "price" in q or "plans" in q:
        return {
            "Basic Plan": KB["pricing"]["Basic Plan"],
            "Pro Plan": KB["pricing"]["Pro Plan"]
        }

    if "basic" in q:
        return KB["pricing"]["Basic Plan"]

    if "pro" in q:
        return KB["pricing"]["Pro Plan"]

    if "refund" in q:
        return KB["policies"]["refund"]

    if "support" in q:
        return KB["policies"]["support"]

    return "Sorry, that information is not available."
```
